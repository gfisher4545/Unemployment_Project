```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import os
import csv
import datetime

output_1973_monthly = 'S&P500/Clean UP/SP500_1973_Monthly_DataFrame.csv'
output_1980_monthly = 'S&P500/Clean UP/SP500_1980_Monthly_DataFrame.csv'
output_2001_monthly = 'S&P500/Clean UP/SP500_2001_Monthly_DataFrame.csv'
output_2008_monthly = 'S&P500/Clean UP/SP500_2008_Monthly_DataFrame.csv'
```

```python
#bring files form csv and format columns
r1973_df = pd.read_csv('S&P500/1968-1978 (Monthly).csv')
r1973_df = r1973_df[['Date', 'Close']]

r1980_df = pd.read_csv('S&P500/1977-1987 (Monthly).csv')
r1980_df = r1980_df[['Date', 'Close']]

r2001_df = pd.read_csv('S&P500/1996-2006 (Monthly).csv')
r2001_df = r2001_df[['Date', 'Close']]

r2008_df = pd.read_csv('S&P500/2003-2013 (Monthly).csv')
r2008_df = r2008_df[['Date', 'Close']]

r2020_df = pd.read_csv('S&P500/2015-Present (Monthly).csv')
r2020_df = r2020_df[['Date', 'Close']]

with open(output_1973_monthly, 'w') as datafile:
    r1973_df.to_csv(output_1973_monthly)

with open(output_1980_monthly, 'w') as datafile:
    r1980_df.to_csv(output_1980_monthly)
    
with open(output_2001_monthly, 'w') as datafile:
    r2001_df.to_csv(output_2001_monthly)

with open(output_2008_monthly, 'w') as datafile:
    r2008_df.to_csv(output_2008_monthly)    
    
r2020_df.head()
```

```python
fig, (ax1,ax2) = plt.subplots(2, sharex=False, figsize=(10,10))
plt.subplots_adjust(hspace=.5)
for ax in [ax1, ax2, ax3, ax4, ax5]:
    ax.set_xlim([0, 130])

ax1.plot(r1973_df['Date'], r1973_df['Close'], linewidth=1, color='c')
ax1.set_ylim(auto=True)
ax1.xaxis.set_major_locator(plt.MaxNLocator(5))
ax1.set_title('1973 Recession')
ax1.xaxis.set_tick_params(rotation=45)
ax1.set_ylabel("S&P 500 Price")
ax1.axvline(x=60,c='black', alpha=0.2)

ax2.plot(r1980_df['Date'], r1980_df['Close'], linewidth=1, color='b')
ax2.set_ylim(auto=True)
ax2.xaxis.set_major_locator(plt.MaxNLocator(5))
ax2.set_title('1980 Recession')
ax2.xaxis.set_tick_params(rotation=45)
ax2.set_ylabel("S&P 500 Price")
ax2.axvline(x=60,c='black', alpha=0.2)

```
