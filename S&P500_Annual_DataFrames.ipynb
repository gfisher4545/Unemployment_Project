```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import os
import csv
import datetime

output_1973_annual = 'S&P500/Clean UP/SP500_1973_Annual_DataFrame.csv'
output_1980_annual = 'S&P500/Clean UP/SP500_1980_Annual_DataFrame.csv'
output_2001_annual = 'S&P500/Clean UP/SP500_2001_Annual_DataFrame.csv'
output_2008_annual = 'S&P500/Clean UP/SP500_2008_Annual_DataFrame.csv'
```

```python
#bring files form csv and format columns
r1973_df = pd.read_csv('S&P500/1968-1978 returns.csv')
r1980_df = pd.read_csv('S&P500/1975-1985 returns.csv')
r2001_df = pd.read_csv('S&P500/1996-2006 returns.csv')
r2008_df = pd.read_csv('S&P500/2003-2013 returns.csv')
r2020_df = pd.read_csv('S&P500/Present returns.csv')


with open(output_1973_annual, 'w') as datafile:
    r1973_df.to_csv(output_1973_annual)

with open(output_1980_annual, 'w') as datafile:
    r1980_df.to_csv(output_1980_annual)
    
with open(output_2001_annual, 'w') as datafile:
    r2001_df.to_csv(output_2001_annual)

with open(output_2008_annual, 'w') as datafile:
    r2008_df.to_csv(output_2008_annual)    
    
r1973_df
```

```python
fig, (ax1,ax2,ax3,ax4, ax5) = plt.subplots(5, sharex=False, figsize=(10,20))
plt.subplots_adjust(hspace=.5)
fig.suptitle("S&P500 Across Recessions", fontsize=16, fontweight="bold")
fig.text(.5, .08, 'Year', ha='center', va='center', fontsize=16)
# for ax in [ax1, ax2, ax3, ax4, ax5]:
#     ax.set_xlim([0, 12])

ax1.plot(r1973_df['date'], r1973_df['return'], linewidth=1, color='c')
ax1.set_ylim(auto=True)
ax1.set_title('1973 Recession')
ax1.xaxis.set_tick_params(rotation=45)
ax1.set_ylabel("S&P 500 Price")

ax2.plot(r1980_df['date'], r1980_df['return'], linewidth=1, color='b')
ax2.set_ylim(auto=True)
ax2.set_title('1980 Recession')
ax2.xaxis.set_tick_params(rotation=45)
ax2.set_ylabel("S&P 500 Price")

ax3.plot(r2001_df['date'], r2001_df['return'], linewidth=1, color='r')
# ax3.xaxis.set_major_locator(plt.MaxNLocator(5))
ax3.set_title('2001 Recession')
ax3.xaxis.set_tick_params(rotation=45)
ax3.set_ylabel("S&P 500 Price")

ax4.plot(r2008_df['date'], r2008_df['return'], linewidth=1, color='g')
# ax4.xaxis.set_major_locator(plt.MaxNLocator(5))
ax4.set_title('2008 Recession')
ax4.xaxis.set_tick_params(rotation=45)
ax4.set_ylabel("S&P 500 Price")

ax5.plot(r2020_df['date'], r2020_df['return'], linewidth=1,color='y')
# ax5.xaxis.set_major_locator(plt.MaxNLocator(5))
ax5.set_title('Present Day')
ax5.xaxis.set_tick_params(rotation=45)
ax5.set_ylabel("S&P 500 Price")

fig.savefig(f'S&P500/Recessions Plot.png')
```
