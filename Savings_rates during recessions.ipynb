```python
import os
import csv
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
savings_rate = "savingsrate.csv"
savings_df = pd.read_csv(savings_rate, encoding="ISO-8859-1")
savings_df.head()
```

```python
savings_df.rename(columns = {"DATE": "Date", "PSAVERT":"Savings_rate"}, inplace=True)
```

```python
savings_df.head()
```

```python
# Create Dataframe for each recession year
# 1973, 1980, 2000, 2008;
# Use index values to reference data
# 5 years +/- or (60mos +/-)

savings_1978 = savings_df.iloc[108:228]
savings_1980 = savings_df.iloc[192:312]
savings_2000 = savings_df.iloc[432:552]
savings_2008 = savings_df.iloc[528:648]
```

```python
savings_2008.head()
```

```python
yaxis_savings = savings_1978['Savings_rate']
xaxis = np.arange(0, 120, 1)
savings1978 = plt.plot(xaxis, yaxis_savings, marker ='o', color='blue')



plt.grid()
plt.title("Savings Rate within 1978 Recession +/-60mos ")
plt.xlabel("Months")
plt.ylabel("Savings Rate")
plt.show()
```

```python
yaxis_savings = savings_1980['Savings_rate']
savings1980 = plt.plot(xaxis, yaxis_savings, marker ='o', color='red')
```

```python
yaxis_savings = savings_2000['Savings_rate']
savings2000 = plt.plot(xaxis, yaxis_savings, marker ='o', color='green')
```
