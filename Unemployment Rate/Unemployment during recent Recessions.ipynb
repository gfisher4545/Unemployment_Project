```python
from matplotlib import pyplot as plt
from scipy import stats
import numpy as np
import pandas as pd
```

```python
unemployment_data = pd.read_csv("unemployement.csv")
recession_1973 = unemployment_data.loc[240:371]
recession_1980 = unemployment_data.loc[324:455]
recession_2001 = unemployment_data.loc[576:707]
recession_2008 = unemployment_data.loc[660:791]
recession_2020 = unemployment_data.loc[804:]

recession_2020
```

```python
recession_1973_high = recession_1973.UNRATE.max()
recession_1980_high = recession_1980.UNRATE.max()
recession_2001_high = recession_2001.UNRATE.max()
recession_2008_high = recession_2008.UNRATE.max()
recession_1973_low = recession_1973.UNRATE.min()
recession_1980_low = recession_1980.UNRATE.min()
recession_2001_low = recession_2001.UNRATE.min()
recession_2008_low = recession_2008.UNRATE.min()
recession_2020_low = recession_2020.UNRATE.min()

print(f"The 1973 Recession had an unemployment rate high of <{recession_1973_high}>")
print(f"The 1980 Recession had an unemployment rate high of <{recession_1980_high}>")
print(f"The 2001 Recession had an unemployment rate high of <{recession_2001_high}>")
print(f"The 2008 Recession had an unemployment rate high of <{recession_2008_high}>")

change_perc_1973 = recession_1973_high/recession_1973_low
change_perc_1980 = recession_1980_high/recession_1980_low
change_perc_2001 = recession_2001_high/recession_2001_low
change_perc_2008 = recession_2008_high/recession_2008_low

predict_1973_unrate = round((change_perc_1973*recession_2020_low),1)
predict_1980_unrate = round((change_perc_1980*recession_2020_low),1)
predict_2001_unrate = round((change_perc_2001*recession_2020_low),1)
predict_2008_unrate = round((change_perc_2008*recession_2020_low),1)

print(f"1973 Recession predicts <{predict_1973_unrate}> Unemployment for 2020")
print(f"1980 Recession predicts <{predict_1980_unrate}> Unemployment for 2020")
print(f"2001 Recession predicts <{predict_2001_unrate}> Unemployment for 2020")
print(f"2008 Recession predicts <{predict_2008_unrate}> Unemployment for 2020")
```

```python
fig, (ax1,ax2,ax3,ax4,ax5) = plt.subplots(5, sharex=False, figsize=(7,20))
fig.subplots_adjust(hspace=.6)
fig.suptitle("Unemployment Rate Across Recessions", fontsize=16, fontweight="bold")
for ax in [ax1, ax2, ax3, ax4, ax5]:
    ax.set_xlim([0, 131])
    
ax1.plot(recession_1973['DATE'], recession_1973['UNRATE'], linewidth=2, color='c')
ax1.set_title('1973 Recession')
ax1.set_ylabel('Unemployment Rate')
ax1.xaxis.set_major_locator(plt.MaxNLocator(5))
ax1.set_ylim([3, 12])
ax1.xaxis.set_tick_params(rotation=45)
ax1.axvline(x=65,c='black', alpha=0.2)

ax2.plot(recession_1980['DATE'], recession_1980['UNRATE'], linewidth=2, color='b')
ax2.set_title('1980 Recession')
ax2.set_ylabel('Unemployment Rate')
ax2.xaxis.set_major_locator(plt.MaxNLocator(5))
ax2.set_ylim([3, 12])
ax2.xaxis.set_tick_params(rotation=45)
ax2.axvline(x=65,c='black', alpha=0.2)

ax3.plot(recession_2001['DATE'], recession_2001['UNRATE'], linewidth=2, color='r')
ax3.set_title('2001 Recession')
ax3.set_ylabel('Unemployment Rate')
ax3.xaxis.set_major_locator(plt.MaxNLocator(5))
ax3.set_ylim([3, 12])
ax3.xaxis.set_tick_params(rotation=45)
ax3.axvline(x=65,c='black', alpha=0.2)

ax4.plot(recession_2008['DATE'], recession_2008['UNRATE'], linewidth=2, color='g')
ax4.set_title('2008 Recession')
ax4.set_ylabel('Unemployment Rate')
ax4.xaxis.set_major_locator(plt.MaxNLocator(5))
ax4.set_ylim([3, 12])
ax4.xaxis.set_tick_params(rotation=45)
ax4.axvline(x=65,c='black', alpha=0.2)

ax5.plot(recession_2020['DATE'], recession_2020['UNRATE'], linewidth=2, color='y')
ax5.set_title('2020 Recession')
ax5.set_ylabel('Unemployment Rate')
ax5.xaxis.set_major_locator(plt.MaxNLocator(5))
ax5.set_ylim([3, 12])
ax5.xaxis.set_tick_params(rotation=45)
ax5.axvline(x=65,c='black', alpha=0.2)

```
