```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
file = 'medianHHincome_monthly.csv'
hh_income_df = pd.read_csv(file)

hh_income_df.columns = ["Date", "Median HH Income"]
hh_income_df
```

```python
tech_bubble = hh_income_df.iloc[132:253]
tech_bubble
```

```python
financial_recession = hh_income_df.iloc[228:349]
financial_recession
```

```python
fig, (ax1, ax2) = plt.subplots(2,figsize = (10,10))
fig.subplots_adjust(hspace = .3)
fig.suptitle("Median Household Income")
ax1.plot(tech_bubble.dropna()['Date'], tech_bubble.dropna()['Median HH Income'], marker = 'o', linestyle = '-', color = 'r')
ax1.set_ylabel('Dollars')
ax1.set_xlabel('Date')
ax1.xaxis.set_major_locator(plt.MaxNLocator(12))
ax2.plot(financial_recession.dropna()['Date'], financial_recession.dropna()['Median HH Income'], linewidth = 1, marker = 'o', color = 'g' )
ax2.set_ylabel('Dollars')
ax2.set_xlabel('Date')
ax2.xaxis.set_major_locator(plt.MaxNLocator(12))
```
