```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import os
import csv
```

Series.to_frame().style

```python
#bring files form csv and format columns
r1973_df = pd.read_csv('S&P500/1968-1978 (Monthly).csv')
r1973_df = r1973_df[['Date', 'Close']]

r1980_df = pd.read_csv('S&P500/1977-1987 (Monthly).csv')
r1980_df = r1980_df[['Date', 'Close']]

r2001_df = pd.read_csv('S&P500/1996-2006 (Monthly).csv')
r2001_df = r2001_df[['Date', 'Close']]

r2008_df = pd.read_csv('S&P500/2003-2013 (Monthly).csv')
r2008_df = r2008_df[['Date', 'Close']]

r2020_df = pd.read_csv('S&P500/2015-Present (Monthly).csv')
r2020_df = r2020_df[['Date', 'Close']]

r2020_df.head()
```

```python
fig, (ax1,ax2,ax3,ax4, ax5) = plt.subplots(5, sharex=False, figsize=(10,20))
plt.subplots_adjust(hspace=.5)
fig.suptitle("S&P500 Across Recessions", fontsize=16, fontweight="bold")
fig.text(.005, .50, 'S&P500 Price', ha='center', va='center', rotation='vertical', fontsize=16)
fig.text(.5, .08, 'Year', ha='center', va='center', fontsize=16)

ax1.plot(r1973_df['Date'], r1973_df['Close'], linewidth=1, marker='o', color='c')
ax1.set_ylim(auto=True)
ax1.xaxis.set_major_locator(plt.MaxNLocator(5))
ax1.set_title('1973 Recession')

ax2.plot(r1980_df['Date'], r1980_df['Close'], linewidth=1, marker='o', color='b')
ax2.set_ylim(auto=True)
ax2.xaxis.set_major_locator(plt.MaxNLocator(5))
ax2.set_title('1980 Recession')

ax3.plot(r2001_df['Date'], r2001_df['Close'], linewidth=1, marker='o', color='r')
ax3.xaxis.set_major_locator(plt.MaxNLocator(5))
ax3.set_title('2001 Recession')

ax4.plot(r2008_df['Date'], r2008_df['Close'], linewidth=1, marker='o', color='g')
ax4.xaxis.set_major_locator(plt.MaxNLocator(5))
ax4.set_title('2008 Recession')

ax5.plot(r2020_df['Date'], r2020_df['Close'], linewidth=1, marker='o', color='y')
ax5.xaxis.set_major_locator(plt.MaxNLocator(5))
ax5.set_title('Present Day')

fig.savefig(f'S&P500/Recessions Plot.png')
```
