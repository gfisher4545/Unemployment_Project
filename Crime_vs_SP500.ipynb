```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import os
import csv
from scipy.stats import linregress
```

```python
#read in crime data
crime_1973 =pd.read_csv('Crime Data/Clean UP/Crime_1973_DataFrame.csv')
crime_1980 =pd.read_csv('Crime Data/Clean UP/Crime_1980_DataFrame.csv')
crime_2001 =pd.read_csv('Crime Data/Clean UP/Crime_2001_DataFrame.csv')
crime_2008 =pd.read_csv('Crime Data/Clean UP/Crime_2008_DataFrame.csv')

#read in SP500 data
sp500_1973 =pd.read_csv('S&P500/Clean UP/SP500_1973_Annual_DataFrame.csv')
sp500_1980 =pd.read_csv('S&P500/Clean UP/SP500_1980_Annual_DataFrame.csv')
sp500_2001 =pd.read_csv('S&P500/Clean UP/SP500_2001_Annual_DataFrame.csv')
sp500_2008 =pd.read_csv('S&P500/Clean UP/SP500_2008_Annual_DataFrame.csv')
```

```python
fig, axes = plt.subplots()

ln1 = axes.plot(sp500_1973['date'], sp500_1973['return'], color='k', label='SP500 Annual Returns (%)')
axes.set_xlabel('Year')
axes.set_ylabel('S&P 500 Annual Returns (%)')
axes.set_title('1973 Recession')
axes.axvline(x=1973,c='black', alpha=0.2)

ax2=axes.twinx()
ln2 = ax2.plot(crime_1973['Year'], crime_1973['Violent Crime Rate'], color='c', label='Crime Rate')
ax2.set_ylabel('Violent Crime Rate per 100,000 Population')

lns = ln1+ln2
labs=[l.get_label() for l in lns]
fig.legend(lns,labs,loc='upper left')
fig.savefig(f'Crime Data/Images/Line - 1973 Crime vs SP500.png')
```
